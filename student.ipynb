```python
X_train_log,X_test_log,Y_train_log,Y_test_log=train_test_split(X,y_log,random_state=0,test_size=0.25)
```

```python
from sklearn.linear_model import LogisticRegression
logistic= LogisticRegression(solver='lbfgs',multi_class='auto',max_iter=1000,random_state=0)
```

```python
logistic.fit(X_train_s,Y_train_log)
```

```python
print("train accuracy after pass fail conversion",logistic.score(X_train_s,Y_train_log))
print("test accuracy after pass fail conversion",logistic.score(X_test_s,Y_test_log))
```
