```python
for i in range(len(y)):
    if(y[i]=='A' or y[i]=="B" or y[i]=='C'):
        y_log[i]="NotF"
    else:
        y_log[i]="F"    
```

```python
from sklearn.model_selection import train_test_split
X_train_log,X_test_log,Y_train_log,Y_test_log=train_test_split(X,y_log,random_state=0,test_size=0.25)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
X_train_s=scaler.fit_transform(X_train_log)
X_test_s=scaler.transform(X_test_log)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier
depth = [5]

grid_param = dict(max_depth=depth)
dtree = GridSearchCV(DecisionTreeClassifier(random_state=0), grid_param, cv=5, iid=False, scoring='roc_auc')
dtree.fit(X_train_s, Y_train_log)

print(dtree.best_params_)

print("test accuracy after pass fail conversion",dtree.score(X_test_s,Y_test_log))

from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
y_pred_tree = dtree.predict(X_test_s)

print(classification_report(Y_test_log,y_pred_tree))
print(confusion_matrix(Y_test_log,y_pred_tree))
```

# Logistic Regression Mulitiple

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,y,random_state=0,test_size=0.20)

from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
logistic= LogisticRegression(solver='lbfgs',multi_class='auto',max_iter=1000,random_state=0)
```

```python
logistic.fit(X_train,Y_train)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, recall_score
y_pred = logistic.predict(X_test)
print(confusion_matrix(Y_test,y_pred))
```

```python
print(classification_report(Y_test,y_pred))
```

```python
print(recall_score(Y_test, y_pred, average='micro'))
```
